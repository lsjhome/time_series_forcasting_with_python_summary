# Chapter 2

# 시계열 예측이란?

시계열 예측은 머신 러닝에서 중요하지만 종종 무시되는 분야이다. 시간이라는 구성이 들어가는 문제가 많기 때문에 중요한 분야이다. 주로 이 시간이라는 문제를 다루기 어려워서 무시되는 경향이 있다. 여기서 시계열 분석에 대해 알아보자. 이 챕터 이후로 당신은:

* 시계열, 시계열 분석, 시계열 예측에 대한 표준 정의를 알게 될 것이다.

* 시계열 데이터에서 생각해야 할 중요한 요소를 알게 될 것이다.

* 시계열에 관한 예시에 대해 알게 될 것이다.

시작해 보자


## 2.1 시계열

보통의 머신 러닝 데이터셋은 관측치들의 모음이다. 예를 들어:

observation #1

observation #2

observation #3

시간은 머신 러닝 데이터셋에서 중요한 역할을 한다. 새로운 데이터에 대한 예측은 실제 결과가 미래에 어떨지 모르는 상황에서 이루어진다. 예측이 이루어 지면서, 모든 이전 데이터는 동등하게 취급이 된다. 아마도 이러한 문제들을 극복하기 위해서 모든 데이터 대신 작년 관측치들만 사용하는 것이 한 방법이 될 수 있다.

시계열 데이터셋은 다르다. 시계열 데이터는 명백한 순서를 관측치 사이에 넣는다: 시간 차원(a time dimension). 이 추가적인 차원은 제약조건이자 추가적인 정보를 주는 구조이다.

시계열이란 시간에 따라 연속적으로 발생한 관측치들의 연속적인 사건들이다. 예를 들어:

Time #1, observation

Time #2, observation

Time #3, observation

## 2.2 시계열 명명법

더 진행하기 전에, 표준적인 용어를 정립하고 시계열 데이터에 대해 알아보자. 현재 시간은 t, 관측 당시 현재 시간은 obs(t)라고 명명된다.

우리는 이전 시간들에 만들어진 관측치들에 대해 관심이 있다. 그리고 그 때 시간들은 lag times 혹은 lags로 불린다. 과거의 시간들은 현재 시간과 음의 관계를 지닌다. 예를 들어 이전 시간이 t-1이고 그 이전 시간은 t-2 이다. 이 시간대의 관측치들은 각각 obs(t-1), obs(t-2)이다. 반대로 미래의 시간은 t+1, t+2 이고 관측치는 각각 obs(t+1), obs(t+2)이다.

간단하게 표기하기 위해, 우리는 종종 obs(t) 표기 대신 t+1을 사용하고 이를 시간 인덱스 대신 사용한다. 추가적으로 시간지연 10(a lag of 10 or lag=10)을 t-10으로 표기한다. 요약하자면:

* t-n : 이전 혹은 지연 시간 (예, t-1는 이전 시간을 말한다.)

* t : 현재 시간 및 기준 시간

* t+n : 미래 혹은 예측 시간 (예, t+1는 다음 시간을 말한다.)

## 2.3 묘사(describing) vs 예측(predicting)

우리는 데이터 셋을 이해하기를 원하느냐 혹은 예측을 하기를 원하느냐에 따라 다른 목표가 있다. 데이터셋을 이해하는 것은, 시계열 분석(time series analysis)라고 불리는데 이는 보다 나은 예측을 하게 도와준다. 하지만 항상 이것이 필요한 것도 아니며 시간과 기술적으로 많은 투자를 필요로 한다. 또한 이것의 결과가 원하는 결과와 잘 일치하지 않을수도 있어서 예측에 부적합할수도 있다.

<p style="text-align: center;">기술(descriptive)적 모델링 혹은 시계열 분석에서, 시계열은(time series) 계절성 패턴, 트렌드들, 외부 요인들 등의 구성 요소들을 결정하기 위해 모델링이 된다. 반대로, 시계열 예측(time series forecasting)은 시계열 안의 정보를 사용하여(아마도 추가적인 정보를 사용하여) 그 시계열의 미래를 예측하는 것이다.</p>

<p style="text-align: right;"> - Practical Time Series Forecasting with R: A Hands-On Guide p.18-19</p>

### 2.3.1 시계열 분석(Time Series Analysis)

전통적인 통계학에서, 주요 쟁점은 시계열 분석이다. 시계열 분석은 기저에 있는 원인을 잘 이해하기 위해 관측된 시계열을 잘 잡아내거나 설명하는 모델을 만드는 것을 포함한다. 이는 종종 데이터의 모양에 대해 가정을 수반하거나 시계열을 여러개의 구성요소들로 분해하는 것을 수반한다. 기술적(descriptive) 모델의 질은 현재 가능한 데이터와 그 해석을 얼마나 잘 제공해서 문제의 영역을 잘 알려주는 것에 달려있다.

<p style="text-align: center;"> 시계열 분석의 주요 목적은 샘플 데이터로부터 타당한 설명을 제공하는 모델을 개발하는 것이다.</p>
<p style="text-align: right;"> - Time Series Analysis and Its Applications: With R examples p.11</p>

### 2.3.2 시계열 예측

미래에 대한 예측은 보외법(extrapolation)이라고 전통적인 통계학에서 불린다. 최근 많은 영역들이 이 주제에 대해 주목하고 이를 시계열 예측이라고 부른다. 예측은 모델을 과거 데이터에 맞게 피팅하는 것과 관측치를 예측하는 것을 수반한다. 기술적 모델들은 미래로부터 빌려오고(모다 더 부드럽게 하거나 노이즈를 제거하기 위해서) 데이터를 가장 잘 설명할 수 있는 방법을 찾는다. 예측에 있어서 중요한 점은, 미래 데이터는 완전히 없고 반드시 이미 있는 데이터로부터 추정(estimated) 되어야 한다는 것이다.

시계열 예측 모델의 기술은 미래를 예측하는 성능에 의해 결정된다. 이 기술은 종종 어째서 이러한 예측이 만들어 졌는지, 신뢰구간은 어떠한지, 그리고 문제에 깔려있는 원인들의 이해를 돕는 원인 들에 대해 알아보는 것을 희생하여 이루어 진다.

## 2.4 시계열의 구성 요소

시계열 분석은 데이터셋을 보다 더 잘 이해하는 기술의 뼈대를 제공한다. 아마도 이중 가장 유용한 방법들 중 하나는 시계열을 4가지 구성 성분으로 나누는 것이다.

* **Level(수준).** 만약 직선일 경우 시간(the series)에 대한 기준치
* **Trend(추세)** 시간을 두고 선택적으로 점차 증가하거나 감소하는 행동들
* **Seasonality(계절성)** 시간을 두고 선택적으로 반복되는 패턴들 혹은 사이클들
* **Noise(소음)** 모델에 의해 설명될 수 없는 선택적인 변수

모든 시계열은 수준(level)이 있고, 대부분 소음이 있으며, 추세와 계절성은 존재할수도 안 할수도 있다.

<p style="text-align: center;"> 많은 시계열의 주요 특징들은 추세들이며 시즌별로 변한다[...]<br>다른 중요한 점은 밀접한 시간을 가진 관측치들이 상관성을 지닐 경향이 있다는 것이다.(연속적으로 종속적이다. serially dependent)</br></p>
<p style="text-align: right;">- Introductory Time Series with R p.2</p>

이를 구성하는 요소들은 관측된 시계열을 제공해 주기 위해 어떤 방식으로 결합될 수 있다. 이 구성요소들의 행동과 어떻게 이들이 결합되느냐에 관한 가정이 만들어 질 수 있으며, 이는 전통적인 통계적 방법을 사용하여서 구성요소들이 모델화 되는 것을 가능하게 해 준다. 이 구성 요소들은 미래에 관한 예측값을 만드는데 가장 효과적인 방법이 될 수도 있지만 항상 그런것은 아니다. 전통적인 통계적 기법들이 뛰어난 효과를 내지 못하는 경우에, 이 구성요소들이 효과적이거나 대체적인 모델들에 대한 입력값이 될 수 있다. 시계열의 수준(level), 추세(trend), 계절성(seasonality), 그리고 소음(noise)로의 분해는 12장에서 더 많이 다루어진다.

## 2.5 예측의 문제점
예측을 할때 중요한 것은 목표를 이해하는 것이다. 소크라테스 방법론을 적용하여 어떠한 문제를 구체적으로 풀기 원하는지 알아보자. 예를 들어서:

1.**얼마나 많은 데이터가 준비되어 있으며 모두 함께 모아두는 것이 가능한가?**
<br>많은 데이터는 보다 도움이 되고, 데이터 탐색, 모델 테스트, 튜닝, 정확도에 관하여 큰 기회기 된다.</br>


2.**필요로 하는 예측의 시간 범위가 어떻게 되는가?**
<br>짧은가, 중간인가, 긴 기간인가? 짧은 기간일수록 높은 신뢰수준으로 예측이 가능하다</br>

3.**예측치들이 시간을 두고 업데이트 되어야 하는가 아니면 한번 만들어 졌으면 정적으로 유지되어야 하는가?**
<br>새 정보가 가능할 때마다 예측을 업데이트 하는 것이 종종 더 정확한 예측을 낳는다</br>

4.**예측을 함에 있어서 어느 수준의 시간주파수가 요구되는가?**
<br>예측들은 높거나 낮은 시간주파수 하에서 만들어진다. 이는 데이터의 업-샘플링 혹은 다운-샘플링을 사용하게 하는데 이는 종종 모델링 과정에 도움을 준다.</br>

시계열 데이터는 종종 클린징, 스케일링, 심지어 변환을 필요로 한다. 예를 들어:
* 주파수(Frequency). 아마도 모델링을 하기에 너무나 높은 주파수로 데이터가 제공되거나 균질하지 않게 간격이 떨어진 데이터가 제공되어 어떠한 모델에서는 리샘플링이 필요할 지도 모른다.

* 이상치(Outliers). 아마도 값들 중에서는 오류가 있거나 극단치를 가진 값이 있어서 식별되거나 처리되어야 할 것이다.

* 결측(Missing). 아마도 보간(interpolate)해야 하거나(중간의 값을 채우다) 결측값을 대체(impute)해야 할 일이 있을 것이다.

종종 시계열 문제는 실시간이고, 계속적으로 예측을 위한 기회제공을 하여야 한다. 이는 모델링을 함에 있어서 나쁜 가정들, 에러들을 빠르게 구체화하여 우리 스스로를 속일 수 있는 시계열 예측에 정직성을 더해준다.

### 2.6 예측의 예시

시계열 예측의 예는 무수히 많다. 아래는 여러 산업에서 시계열 분석과 예측을 중요하게 여기는 10가지 예다.

* 몇 톤이나 옥수수를 각 주에서 생산할 것인지 예측

* 초단위의 EEG(수면뇌파)가 환자가 발작을 일으킬 것인지를 나타내는가 예측

* 매일 주가의 마감 가격 예측

* 한 도시에서 모든 병원들에서 출산율이 매년 어떻게 될 것인지 예측

* 한 가게에서 매일 몇개의 상품이 팔릴 것인지 예측

* 매일 역을 지나는 사람들의 수를 예측

* 각 분기마다 주별로 실업율 예측

* 각 시간대 마다 서버의 이용도 예측

* 각 주의 산란기마다 토끼 인구의 크기 예측

* 도시에서 매일 평균 가소린 가격이 얼마나 될 것인지 예측

이쯤하면 스스로 풀어보고 싶은 시계열 문제를 떠올릴 수 있을 것이다.
